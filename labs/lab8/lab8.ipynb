### Зв'язок ціни на нафту з ppp

```python
brent_year['Date'] = pd.to_datetime(brent_year['Date'])
brent_year['Year'] = brent_year['Date'].dt.year

merged_data = brent_year.merge(ppp, left_on='Year', right_on='Year', how='inner')

plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Price'], merged_data['PPP'], alpha=0.5)
plt.title('Зв\'язок між ціною на нафту та PPP')
plt.xlabel('Ціна на нафту')
plt.ylabel('PPP')
plt.grid(True)

plt.show()
```

```python
sns.lmplot(x='Price', y='PPP', data=merged_data, height=6)
plt.title('Зв\'язок між ціною на нафту та PPP')

plt.xlabel('Ціна на нафту')
plt.ylabel('PPP')

plt.show()
```

### Відсоток ppp окремої країни до серднього ppp всіх країн за рік

```python
merged_data = population.merge(ppp, left_on=['Country Name', 'Year'], right_on=['Country', 'Year'], how='inner')

average_ppp_by_year = merged_data.groupby('Year')['PPP'].mean()

merged_data['Average PPP'] = merged_data['Year'].map(average_ppp_by_year)
merged_data['PPP Percentage'] = (merged_data['PPP'] / merged_data['Average PPP']) * 100

merged_data
```

```python
merged_data[(merged_data['Country Name'] == 'India')]
```

```python
merged_data[(merged_data['Country Name'] == 'India')].plot(y='PPP Percentage', x='Year')
```

```python
merged_data[(merged_data['Country Name'] == 'Trinidad and Tobago')]
```

```python
merged_data[(merged_data['Country Name'] == 'Trinidad and Tobago')].plot(y='PPP Percentage', x='Year')
```

### Зв'язок популяції та ppp

```python
merged_data = population.merge(ppp, left_on=['Country Name', 'Year'], right_on=['Country', 'Year'], how='inner')

plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Value'], merged_data['PPP'], alpha=0.5)
plt.title('Зв\'язок між популяцією та PPP')
plt.xlabel('Популяція')
plt.ylabel('PPP')
plt.grid(True)

plt.show()
```

```python
sns.lmplot(x='Value', y='PPP', data=merged_data, height=6)
plt.title('Зв\'язок між популяцією та PPP з лінією регресії')

plt.xlabel('Популяція')
plt.ylabel('PPP')

plt.show()
```
